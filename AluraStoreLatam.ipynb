### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
def obtenerIngreso(lista: list):
  return sum(lista)
```

```python
ingreso_tienda_1 = obtenerIngreso(tienda['Precio'])
ingreso_tienda_2 = obtenerIngreso(tienda2['Precio'])
ingreso_tienda_3 = obtenerIngreso(tienda3['Precio'])
ingreso_tienda_4 = obtenerIngreso(tienda4['Precio'])
```

# 2. Ventas por categoría

```python
producto_categoria_tienda_1 = list(tienda['Categoría del Producto'].value_counts().items())
producto_categoria_tienda_2 = list(tienda2['Categoría del Producto'].value_counts().items())
producto_categoria_tienda_3 = list(tienda3['Categoría del Producto'].value_counts().items())
producto_categoria_tienda_4 = list(tienda4['Categoría del Producto'].value_counts().items())
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda_1 = sum(tienda['Calificación'])/len(tienda['Calificación'])
promedio_calificacion_tienda_2 = sum(tienda2['Calificación'])/len(tienda2['Calificación'])
promedio_calificacion_tienda_3 = sum(tienda3['Calificación'])/len(tienda3['Calificación'])
promedio_calificacion_tienda_4 = sum(tienda4['Calificación'])/len(tienda4['Calificación'])
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_tienda_1 = list(tienda['Producto'].value_counts().items())
productos_vendidos_tienda_1.sort(key=lambda x: x[1], reverse=True)
productos_vendidos_tienda_2 = list(tienda2['Producto'].value_counts().items())
productos_vendidos_tienda_2.sort(key=lambda x: x[1], reverse=True)
productos_vendidos_tienda_3 = list(tienda3['Producto'].value_counts().items())
productos_vendidos_tienda_3.sort(key=lambda x: x[1], reverse=True)
productos_vendidos_tienda_4 = list(tienda4['Producto'].value_counts().items())
productos_vendidos_tienda_4.sort(key=lambda x: x[1], reverse=True)
```

# 5. Envío promedio por tienda

```python
costo_envio_tienda_1 = sum(tienda['Costo de envío'])/len(tienda['Costo de envío'])
costo_envio_tienda_2 = sum(tienda2['Costo de envío'])/len(tienda2['Costo de envío'])
costo_envio_tienda_3 = sum(tienda3['Costo de envío'])/len(tienda3['Costo de envío'])
costo_envio_tienda_4 = sum(tienda4['Costo de envío'])/len(tienda4['Costo de envío'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.bar(labels, ventas, color = colors)
plt.title(' Valor de Ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Valor de Ventas')
for i, value in enumerate(ventas):
    plt.text(i, value, str(round(value, 2)), ha='center')
plt.show()
```

```python
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valor = [costo_envio_tienda_1, costo_envio_tienda_2, costo_envio_tienda_3, costo_envio_tienda_4]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.bar(labels, valor, color = colors)
plt.title('Costo de envio promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envio Promedio')
for i, value in enumerate(valor):
    plt.text(i, value, str(round(value, 2)), ha='center')
plt.show()
```

```python
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valor = [(ingreso_tienda_1 - sum(tienda['Costo de envío'])), (ingreso_tienda_2 - sum(tienda2['Costo de envío'])), (ingreso_tienda_3 - sum(tienda3['Costo de envío'])), (ingreso_tienda_4 - sum(tienda4['Costo de envío']))]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.bar(labels, valor, color = colors)
plt.title('Ingreso neto por tienda (ventas totales - costos de envio))')
plt.xlabel('Tienda')
plt.ylabel('Ingreso neto por tienda (ventas totales - costos de envio))')
for i, value in enumerate(valor):
    plt.text(i, value, str(round(value, 2)), ha='center')
plt.show()
```

```python
labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
valor = [promedio_calificacion_tienda_1, promedio_calificacion_tienda_2, promedio_calificacion_tienda_3, promedio_calificacion_tienda_4]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.bar(labels, valor, color = colors)
plt.title('Promedio de calificacion')
plt.xlabel('Tienda')
plt.ylabel('Promedio de calificacion')
for i, value in enumerate(valor):
    plt.text(i, value, str(round(value, 2)), ha='center')
plt.show()
```

```python
dic = {1:producto_categoria_tienda_1,
       2:producto_categoria_tienda_2,
       3:producto_categoria_tienda_3,
       4:producto_categoria_tienda_4}

fig1 = plt.figure("Ventas por categoria", figsize=(12, 8))
fig1.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 5):
    lista = dic[i]
    labels = [label[0] for label in lista]
    valor = [label[1]/sum([item[1] for item in lista]) for label in lista]
    ax = fig1.add_subplot(2, 2, i)
    ax.pie(valor, labels=labels, autopct='%.1f%%', startangle=90, wedgeprops={"edgecolor": "white"})
    ax.set_title('Ventas por categoria tienda ' + str(i))

plt.show()

```

# **Informe**

El presente informe tiene como objetivo determinar cuál de las tiendas del Sr. Juan presenta el rendimiento más bajo en función de diversos indicadores clave de desempeño, con el fin de evaluar si debe ser vendida o cerrada. Este análisis considera datos cuantitativos relacionados a ventas, ingresos, productos, satisfacción de clientes y costos logísticos, y se apoya en gráficos y visualizaciones generadas para respaldar la decisión final.

📈 Desarrollo del Análisis
1. Ingresos Totales por Tienda
Al evaluar los ingresos totales generados por cada tienda, se observa que la Tienda 4 presenta los valores más bajos respecto a sus pares. Las tiendas 1, 2 y 3 tienen un rendimiento significativamente superior, lo cual indica una mayor estabilidad y rentabilidad.

2. Categorías de Productos Más y Menos Vendidas
Los gráficos de distribución de ventas por categoría revelan que las categorías más vendidas en general son Muebles, Electrónica y Juguetes. Aunque la Tienda 4 también presenta ventas en estas categorías, su volumen total es reducido en comparación con las demás tiendas, lo cual sugiere que no logra capitalizar los productos de alto movimiento.

3. Calificaciones Promedio de los Clientes
En cuanto a la satisfacción del cliente, la Tienda 4 no destaca negativamente. Su calificación promedio se encuentra por encima de una tienda, pero por debajo de las otras dos, lo que indica un desempeño aceptable en servicio, aunque no sobresaliente.

4. Productos Más y Menos Vendidos
Se analizaron los productos individuales más vendidos por tienda, y nuevamente la Tienda 4 aparece con menor volumen, incluso en productos de alta demanda. Además, presenta una mayor proporción de productos con baja rotación, lo cual afecta su eficiencia operativa.

5. Coste de Envío Promedio
Un punto clave es el coste de envío promedio por tienda. La Tienda 4 presenta un costo logístico más elevado por unidad vendida, probablemente debido al bajo volumen de ventas y la baja densidad de envíos, lo que reduce su competitividad y margen de beneficio.

✅ Conclusión y Recomendación
Recomendación: Vender o cerrar la Tienda 4.

Justificación:

Es la tienda con menores ingresos totales.

Tiene un menor volumen de ventas, incluso en las categorías más populares.

Su calificación de clientes es aceptable, pero no suficiente para compensar su bajo rendimiento económico.

Los costes de envío son altos, reduciendo aún más su rentabilidad.

En contraste, las tiendas 1, 2 y 3 muestran mejores métricas en todos los aspectos analizados.

Por estas razones, se concluye que la Tienda 4 es la opción más viable para ser vendida. No obstante, se recomienda reubicar a los empleados con mejor desempeño a otras tiendas para conservar el talento valioso.
